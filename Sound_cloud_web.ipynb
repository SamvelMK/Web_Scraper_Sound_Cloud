```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import pandas as pd

# from lxml import html
# import requests
```

### Data we want
* Item = div class chartTracks__item
* Rating = div class = chartTrack__position sc-font-tabular
* Name =  div chartTrack__username sc-type-light sc-truncate
* KB = span sc-visuallyhidden

```python
# Setting up a driever object with Chrome
driver = webdriver.Chrome('pathdriver.set_page_load_timeout(10)
driver.get('https://soundcloud.com/charts/top')

# Accepts the cookies and closes the announcment bar on the webpage
driver.find_element_by_xpath("//button[@class='announcement__ack sc-media-right sc-button sc-button-medium']").click()
driver.find_element_by_xpath("//a[@class='announcement__dismiss']").click()

# Scrols down to get the full html
SCROLL_PAUSE_TIME = 0.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height   

# Gets the html and quits the webpage    
html = driver.page_source
driver.quit()
```

```python
# Parses the html file and extracts the segment of HTML that we contains all our information
soup = BeautifulSoup(html, 'html.parser')
all = soup.find_all('div', {'class':'chartTracks lazyLoadingList'})
```

```python
# extracting information we need and puting it in a dictionary and attaching it to the empaty list
l = []
for i in range(0,50):
    d = {}
    d['Memory in KB'] = all[0].find_all('span', {'class': 'sc-visuallyhidden'})[i].text
    d['Name'] = all[0].find_all('a', {'class':'sc-link-dark'})[i].text
    d['Ranking'] = all[0].find_all('div', {'class':'chartTrack__position sc-font-tabular'})[i].text
    d['Owner'] = all[0].find_all('a', {'class': 'sc-link-light'})[i].text
    l.append(d)
```

```python
# Converting a list of dictionary object to pandas DF
df = pd.DataFrame(l)

# Changing the order of the columns
data = df[['Ranking', 'Name', 'Owner', 'Memory in KB']]
```
